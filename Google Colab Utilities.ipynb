```python
# @title Citation Parser
import pandas as pd
from google.colab import files
def parse_citation(multiline_string):
    # create a pandas dataframe to store the data
    df = pd.DataFrame(columns=['Title', 'Authors', 'Journal', 'DOI', 'Year'])
    m_string = multiline_string.split('\n')
    m_string = m_string[1:-1]
    for line in m_string:
        # for link like reference
        if "[Online]" in line:
            first_https = line.index("https://")
            doi = line[first_https:]
            first_quote = line.index('“')
            last_quote = line.index('”')
            main_body = line[first_quote+1:last_quote].strip()[:-1]
            if '|' in main_body:
                first_pipe = main_body.index('|')
                title = main_body[:first_pipe]
                authors = ""
                journal = main_body[first_pipe+1:]
                year = ""
            elif 'Wiley Online Library' in main_body:
                main_body = main_body.split(' - ')
                title = main_body[0]
                authors = main_body[1]
                journal = main_body[2]
                year = main_body[3]
            else:
                title = main_body
                authors = ""
                journal = ""
                year = ""
        # for normal reference
        else:
            # find the index of occurance of first "
            first_end_sq = line.index(']')
            first_quote = line.index('“')
            authors = line[first_end_sq+1:first_quote].strip()[:-1]
            # print(authors)
            last_quote = line.index('”')
            title = line[first_quote+1:last_quote].strip()[:-1]
            # print(title)
            frist_comma_after_end_quote = line.index(',', last_quote)
            journal = line[last_quote+1:frist_comma_after_end_quote].strip()
            # print(journal)
            first_doi = line.index('doi:')
            doi = 'https://doi.org/'+line[first_doi+4:].strip()[:-1]
            # print(doi)
            line_list = line.split(',')
            year = line_list[-2].strip()
            # print(year)

        item = [title, authors, journal, doi, year]
        df.loc[len(df)] = item

    return df
```
